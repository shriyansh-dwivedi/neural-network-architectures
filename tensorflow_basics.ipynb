# 2. Perform basic tensor operations including matrix addition, multiplication, and transpose.

# Addition

```python
add_result = A+B
print('Addition:',add_result)
```

# Multiply

```python
mul_result = torch.matmul(A,B)
print('Multiply:',mul_result)
```

# Transpose

```python
transpose_A = A.T
print('Transpose:',transpose_A)
```

# 3. Convert the resulting tensor into a Numpy array and then reshape it into a (1,9) tensor.

```python
import torch
A = torch.rand(3,3)
B = torch.rand(3,3)

print('Tensor:',A)
print('Tensor:',B)
```

```python
C = A + B
print('Result Addition:',A)
```

```python

C_numpy = C.numpy()
print('Converted to NumPy Array:',C_numpy)
```

```python
C_reshaped = C_numpy.reshape(1, 9)
print('Reshaped NumPy Array:',C_reshaped)
```
